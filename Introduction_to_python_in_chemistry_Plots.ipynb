```python
import matplotlib.pyplot as plt
```

```python
# Get the atomic radius (y-axis) data to plot against atomic number (x-axis)
number = ptable.atomic_number
radius = ptable["atomic_radius"]
print(number,radius)
```

```python
# Line plot
plt.plot(number,radius)
```

```python
# Change color
plt.plot(number,radius)
```
